```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
os.chdir(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\03_OBS\analysis_3\infrastructure_exposure\csv_forroadrasters_withdepthstats_present')

da = pd.read_csv('floodedroadareas_compound.csv', index_col=0)
da.set_index(keys='County', drop=True, inplace=True)
pres_comp = da[['MedianDepth(m)']]
pres_comp.columns = ['Compound']

da = pd.read_csv('floodedroadareas_runoff.csv', index_col=0)
da.set_index(keys='County', drop=True, inplace=True)
pres_run = da[['MedianDepth(m)']]
pres_run.columns = ['Runoff']

da = pd.read_csv('floodedroadareas_coastal.csv', index_col=0)
da.set_index(keys='County', drop=True, inplace=True)
pres_coast = da[['MedianDepth(m)']]
pres_coast.columns = ['Coastal']

da = pd.read_csv('floodedroadareas_alldrivers.csv', index_col=0)
da.set_index(keys='County', drop=True, inplace=True)
pres_comb = da[['MedianDepth(m)']]
pres_comb.columns = ['Total']

present_data = pd.concat(objs=[pres_run, pres_coast, pres_comp, pres_comb], axis=1, ignore_index=False)
present_data.dropna(axis=0, how='all', inplace=True)
print(present_data)
```

```python
scenarios = ['Runoff', 'Coastal', 'Compound', 'Total']

for i in range(len(scenarios)):
    pres = present_data[scenarios[i]]
    d1 = pres[~pres.isna()].values
    print(scenarios[i], np.median(d1))
```
