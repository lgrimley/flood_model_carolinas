```python
# All runs
df = pd.read_csv(r'.\fldArea_by_process.csv', index_col=0)
df = cleanup_flood_area_dataframe(df)

df_p = df[df['climate'] == 'pres']
df_p.drop(columns = ['storm', 'climate', 'group'], inplace=True)

df_f = df[df['climate'] == 'fut']
df_f.drop(columns = ['storm', 'climate', 'group'], inplace=True)

diff = np.round(df_f-df_p.iloc[0], 2)
print('change in total flooded area')
print(diff.describe())

print('% change in total flooded area (relative to the present)')
reldiff = np.round(diff/df_p.iloc[0], 2)*100
print(reldiff.describe())
```

```python
for storm in df['storm'].unique():
    print(storm)
    dfs = df[df['storm'] == storm]
    df_p = dfs[dfs['climate'] == 'pres']
    df_p.drop(columns = ['storm', 'climate', 'group'], inplace=True)
    
    df_f = dfs[dfs['climate'] == 'fut']
    df_f.drop(columns = ['storm', 'climate', 'group'], inplace=True)

    diff = np.round(df_f-df_p.iloc[0], 2)
    print('change in total flooded area')
    print(diff.describe())
    
    print('% change in total flooded area (relative to the present)')
    reldiff = (diff/df_p.iloc[0])*100
    print(reldiff.describe())
```

```python
# Calculate fractional contribution
for scen in ['Coastal', 'Runoff', 'Compound']:
    df[f'{scen}_RelCont'] = df[scen] / df['Total']

# Storm stats
df_rc = df[['Coastal_RelCont', 'Runoff_RelCont', 'Compound_RelCont', 'group']]
stats_rc =  np.round(df_rc.groupby('group').agg(['mean', 'std', 'median'])*100, decimals=2)
print(stats_rc)

# Storm average stats
df_rc = df[['Coastal_RelCont', 'Runoff_RelCont', 'Compound_RelCont', 'climate']]
stats_rc2 = np.round(df_rc.groupby('climate').agg(['mean', 'std', 'median'])*100, decimals=2)
print(stats_rc2)
```
