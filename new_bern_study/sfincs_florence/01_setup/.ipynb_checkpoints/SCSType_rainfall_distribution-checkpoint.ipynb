```python
'''
Date: 8/29/2024
Author: Lauren Grimley
Contact: lauren.grimley@unc.edu

Description: 
This script is an overview of how to take a 24-hour storm total precipitation (e.g., from NOAA Atlas 14) 
and apply a SCS Type distribution to get a storm hyetograph

Consider learning more about what the SCS type curves are and where they came from before applying. 
Here are some starting resources:
- https://www.hec.usace.army.mil/confluence/hmsdocs/hmstrm/meteorology/precipitation/hypothetical-storm
- https://hdsc.nws.noaa.gov/pfds/
'''
```

```python
import os
import urllib.request
import re
import pandas as pd
```

```python
scsTypeCurves = pd.DataFrame()
for i in ['I', 'II', 'III']:
    path = f'https://www.hydrocad.net/rainfall/tables/Type%20{i}%2024-hr%20Tabular.hcr'
    f = urllib.request.urlopen(path)
    fd = f.read()
    fd = fd.decode("utf-8")
    lines = fd.split('\n')
    # Read the data from the webpage to get the SCS distribution
    dist = []
    for line in lines:
        if 'depth=' in line:
            data = re.findall(r'[.]?\d+[.]?\d+', line)
            data_int = [float(x) for x in data]
            dist = dist + data_int

    dist = pd.DataFrame(dist)
    dist.columns = [f'SCSType{i}']
    scsTypeCurves = pd.concat([scsTypeCurves, dist], axis=1)

# Setup the time component of the distribution
scsTypeCurves['time_hr'] = scsTypeCurves.index/10.0
scsTypeCurves.set_index('time_hr', inplace=True, drop=True)

# Print the top of the dataframe with the distributions
print(scsTypeCurves.head(10))
```

```python
# Plot the SCS type curves
scsTypeCurves.plot(xlabel = 'Time (hours)', ylabel = 'Fraction of 24-hour rainfall (in)')
```

```python
# Set the NOAA Atlas design storm total in INCHES
total_precip = 10
# Map this to the SCS distributions to get the fractional rainfall over the 24 hour storm
# Because this is the cummulative rainfall over the storm, the final row should equal the total above 
cumm_precip = scsTypeCurves * total_precip
print(cumm_precip.tail(4))
```

```python
# Now we want to get the rain rate from the cummulative rainfall over the storm
rain_rate = cumm_precip.diff(periods=1)

# We can see what the peak 6-min rain rate is for each SCS distribution
print('Max rain rate (in/6min)\n', rain_rate.max())

# If we sum the rain rates, we will get the original precipitation total
print('\nTotal 24-hour rainfall (in)\n', rain_rate.sum())
```

```python
# Plot the hyetographs
rain_rate.plot(xlabel = 'Time (hours)', ylabel = 'Rain Rate (inches / 6-min)')
```

```python
# We can resample to an hourly rain rate
rain_rate['time'] = pd.to_timedelta(rain_rate.index, unit='hour')
rain_rate.set_index('time', inplace=True, drop=True)
rain_rate = rain_rate.resample(rule='h').sum()

# Let's see what the hourly peak rain rate is...
print('Max rain rate (in/hr)\n', rain_rate.max())

# If we sum the rain rates, we will get the original precipitation total
print('\nTotal 24-hour rainfall (in)\n', rain_rate.sum())
```
