```python
import pandas as pd
import matplotlib.pyplot as plt

build = pd.read_csv(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\florence_pres_fut_buildings.csv')
print(build.head(3))
```

```python
build['depth_diff'] = build['fut_depth'] - build['pres_depth']


fig, axes = plt.subplots(nrows=3, ncols=1, sharey=False, sharex=False)
build['pres_depth'].plot(kind='hist',bins=25, ax=axes[0], density=True)
axes[0].set_xlim(0,5)
build['fut_depth'].plot(kind='hist', bins=25, ax=axes[1], density=True)
axes[1].set_xlim(0,5)
build['depth_diff'].plot(kind='hist', bins=25, ax=axes[2], density=True)
axes[2].set_xlim(0,1.5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharey=True, sharex=True)
build['pres_vmax'].plot(kind='hist',bins=25, ax=axes[0], density=True)
build['fut_vmax'].plot(kind='hist', bins=25, ax=axes[1], density=True)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharey=True, sharex=True)
build['pres_tmax'].plot(kind='hist',bins=25, ax=axes[0], density=True)
build['fut_tmax'].plot(kind='hist', bins=25, ax=axes[1], density=True)
```
