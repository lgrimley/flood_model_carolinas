```python
#!/usr/bin/env python
# coding: utf-8

import os
import re
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import datetime as dt
import xarray as xr
import matplotlib as mpl
import hydromt
from hydromt import DataCatalog
import hydromt_sfincs
from hydromt_sfincs import SfincsModel
```

```python
# Filepath to data catalog yml
cat_dir = r'Z:\Data-Expansion\users\lelise\data'
yml_base_CONUS = os.path.join(cat_dir, 'data_catalog_BASE_CONUS.yml')
yml_base_Carolinas = os.path.join(cat_dir, 'data_catalog_BASE_Carolinas.yml')
yml_sfincs_Carolinas = os.path.join(cat_dir, 'data_catalog_SFINCS_Carolinas.yml')
root = r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\01_AGU2023\future_florence\future_florence_ensmean'
mod = SfincsModel(root=root, mode='r', data_libs=[yml_base_CONUS, yml_base_Carolinas, yml_sfincs_Carolinas])
cat = mod.data_catalog
studyarea_gdf = mod.region.to_crs(epsg=32617)
studyarea_gdf.plot()
```

```python
''' Read in NC and SC buildings data '''
# Read in structures information and clip to the study area. This might take a little while and only needs to be run once...
nc_buildings = gpd.read_file(filename=r'Z:\Data-Expansion\users\lelise\data\storm_data\hurricanes\X_observations\nfip_flood_damage_NC'
                             r'\included_data.gdb',
                             layer='buildings',
                             mask=studyarea_gdf).to_crs(studyarea_gdf.crs)
nc_buildings['STATE'] = 'NC'
b1 = nc_buildings.drop(nc_buildings.columns[~nc_buildings.columns.isin(['STATE', 'geometry'])], axis=1)
print('Number of NC Buildings in Study Area:', str(len(nc_buildings)))

# Load SC buildings from NSI. This might take a little while and only needs to be run once...
sc_buildings = gpd.read_file(filename=r'Z:\Data-Expansion\users\lelise\data\geospatial\infrastructure\nsi_2022_45.gpkg',
                             mask=studyarea_gdf).to_crs(studyarea_gdf.crs)
sc_buildings['STATE'] = 'SC'
b2 = sc_buildings.drop(sc_buildings.columns[~sc_buildings.columns.isin(['STATE', 'geometry'])], axis=1)
print('Number of SC Buildings in Study Area:', str(len(sc_buildings)))

# Combine NC and SC data into single dataframe
buildings = pd.concat(objs=[b1, b2], axis=0, ignore_index=True)
print('Number of Buildings in Study Area:', str(len(buildings)))
```

```python
# Load flood model data
os.chdir(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\03_OBS\analysis_3')
da_zsmax = xr.open_dataset('pgw_zsmax.nc', engine='netcdf4')
da_vmax = xr.open_dataset('pgw_vmax.nc', engine='netcdf4')
da_tmax = xr.open_dataset('pgw_tmax.nc', engine='netcdf4')

# Get hurricane Florence
run_ids = da_zsmax.run.values
subset_list = [r for r in run_ids if 'flor_fut' in r]
subset_list = [r for r in subset_list if 'compound' in r]
subset_list = [r for r in subset_list if 'SF8' not in r]
print(subset_list)
```

```python
''' PART 2 - Extract flood depths and process attribution '''
gdf = buildings.copy()
gdf['xcoords'] = gdf['geometry'].x.to_xarray()
gdf['ycoords'] = gdf['geometry'].y.to_xarray()
```

```python
# Get the peak WSE for the present
da_zsmax_pres = da_zsmax.sel(run='flor_pres_compound')
v = da_zsmax_pres['zsmax'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
gdf[f'pres_zsmax'] = v.transpose()

# Get the ground elevation
dep = xr.open_dataset(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\subgrid\dep_subgrid_20m.tif')
v = dep['band_data'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
gdf['gnd_elev'] = v.transpose()
```

```python
# Get the mean WSE for the future ensemble
#da_zsmax_fut = da_zsmax.sel(run=subset_list).mean(dim='run')
#v = da_zsmax_fut['zsmax'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
#gdf[f'fut_zsmax'] = v.transpose()

da_zsmax_ensmean = xr.open_dataset(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\03_OBS\analysis_3\ensemble_mean\fut_ensemble_zsmax_mean.nc')
da_zsmax_fut = da_zsmax_ensmean.sel(run='flor_fut_compound_mean')
#da_zsmax_fut['flor_fut_coastal_mean'].raster.to_raster(r'Z:\Data-Expansion\users\lelise\data_share\SFINCS_OUTPUT\pgw_version_20240720\max_water_level_meters\flor_fut_compound_ensmean_peakWL.tif')

#v = da_zsmax_fut['flor_fut_coastal_mean'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
#gdf[f'fut_zsmax'] = v.transpose()
```

```python
# Calculate the peak water depth
gdf['fut_depth'] = gdf['fut_zsmax'] - gdf['gnd_elev']
gdf['pres_depth'] = gdf['pres_zsmax'] - gdf['gnd_elev']

# Remove buildings that did not flood
gdf = gdf[(gdf['fut_depth'] > 0) | (gdf['pres_depth'] > 0)]
print(len(gdf))
```

```python
# Get the flood process classification from the future ensemble mean
da_zsmax_ensmean_class = xr.open_dataset(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\03_OBS\analysis_3\ensemble_mean\processes_classified_ensmean_mean.nc')
da_zsmax_ensmean_class = da_zsmax_ensmean_class.sel(run='flor_fut_ensmean')
v = da_zsmax_ensmean_class['flor_fut_ensmean'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
gdf[f'fut_zsmax_class'] = v.transpose()
```

```python
# Calculate the depth difference
gdf['depth_diff'] = gdf['fut_depth'] - gdf['pres_depth']
print(gdf['depth_diff'].describe())
```

```python
import matplotlib.dates as mdates
from matplotlib.colors import LinearSegmentedColormap
from datetime import datetime, timedelta
from mpl_toolkits.axes_grid1 import make_axes_locatable
import cartopy.crs as ccrs

thresh = 0.5
da_plot = gdf[gdf['depth_diff'] > thresh]
vmin=thresh
vmax=2
colormap='Blues'

# MAP THE STATS
da = mod.grid['dep']
wkt = da.raster.crs.to_wkt()
utm_zone = da.raster.crs.to_wkt().split("UTM zone ")[1][:3]
utm = ccrs.UTM(int(utm_zone[:2]), "S" in utm_zone)

font = {'family': 'Arial', 'size': 10}
mpl.rc('font', **font)
mpl.rcParams.update({'axes.titlesize': 10})

fig, ax = plt.subplots(
    nrows=1, ncols=1,
    figsize=(6.2, 4.2),
    subplot_kw={'projection': utm},
    tight_layout=True
)

mod.region.plot(ax=ax, color='lightgray', edgecolor='none', linewidth=0, linestyle='-', zorder=2, alpha=0.75)
da_plot.plot(column='depth_diff',
                cmap=colormap,
                legend=False,
                vmin=vmin, vmax=vmax,
                ax=ax,
                markersize=15,
                edgecolor='black',
                linewidth=0.25,
                zorder=3
                )

#  Setup colorbar and add to plot
sm = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
fig.colorbar(sm,
             ax=ax,
             shrink=0.9,
             label='m',
             extend='max',
             spacing='uniform')
```

```python
# Get the flood process classification from the future ensemble mean
da_zsmax_ensmean_class = xr.open_dataset(r'Z:\Data-Expansion\users\lelise\projects\Carolinas_SFINCS\Chapter2_PGW\sfincs\03_OBS\analysis_3\ensemble_mean\processes_classified_ensmean_mean.nc')
da_zsmax_ensmean_class = da_zsmax_ensmean_class.sel(run='flor_fut_ensmean')
v = da_zsmax_ensmean_class['flor_fut_ensmean'].sel(x=gdf['geometry'].x.to_xarray(), y=gdf['geometry'].y.to_xarray(), method='nearest').values
gdf[f'fut_zsmax_class'] = v.transpose()
```

```python
compound = gdf[(gdf['fut_zsmax_class'] == 2) | (gdf['fut_zsmax_class'] == 4)]
print(compound['depth_diff'].describe())

coastal = gdf[(gdf['fut_zsmax_class'] == 1)]
print(coastal['depth_diff'].describe())

runoff = gdf[(gdf['fut_zsmax_class'] == 3)]
print(runoff['depth_diff'].describe())
```

```python
thresh = -75
da_plot = [runoff, coastal, compound]
vmin=thresh
vmax=75
colormap='bwr'
var = 'tmax_diff'
unit = 'Difference in Time of Inundation (hr)'

fig, axs = plt.subplots(
    nrows=1, ncols=3,
    figsize=(9, 4.5),
    subplot_kw={'projection': utm},
    tight_layout=True
)
axs = axs.flatten()
for i in range(len(da_plot)):
    ax = axs[i]
    d = da_plot[i]
    d = d[d[var] > thresh]
    mod.region.plot(ax=ax, color='lightgray', edgecolor='none', linewidth=0, linestyle='-', zorder=2, alpha=0.75)
    d.plot(column=var,
                    cmap=colormap,
                    legend=False,
                    vmin=vmin, vmax=vmax,
                    ax=ax,
                    markersize=10,
                    edgecolor='black',
                    linewidth=0.15,
                    zorder=3,
           alpha=0.7
                    )

#  Setup colorbar and add to plot
sm = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
fig.colorbar(sm,
             ax=axs[2],
             shrink=0.7,
             label=unit,
             extend='both',
             spacing='uniform')

plt.subplots_adjust(wspace=0, hspace=0)
plt.margins(x=0, y=0)
```
